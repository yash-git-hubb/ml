```python
def function(x):
    return (x+3)**2

def derivative(x):
    return 2*(x+3)

def grad():
    x = 2  
    learning_rate = 0.1
    num_iterations = 100
    
    for i in range(num_iterations):
        x = x - learning_rate*derivative(x)
        y = function(x)
        print(f"Iteration {i+1}: x = {x}, y = {y}")
    print(f"The local minimum occurs at x = {x}, y = {y}")
```

```python
grad()
```
